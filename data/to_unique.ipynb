Makes a version of given dataset with only unique words

```python
import pandas as pd

data_file = 'kidwords_5000000_020724.csv'
data_folder = '.'
data_path = data_folder + '/' + data_file

data = pd.read_csv(data_path)
size = data.shape[0]

def to_unique(dataset):
    return dataset.drop_duplicates()

data = to_unique(data)

print('Dataset changed from', size, 'to', data.shape[0], 'rows')
print('Saving to', data_path[:-4] + '_unique.csv')
data.to_csv(data_path[:-4] + '_unique.csv', index=False)
```
