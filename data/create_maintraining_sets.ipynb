```python
import pandas as pd
from preprocess import main
import numpy as np
into_reading = pd.read_csv("into_reading/into_reading_sequentially_arranged_repeated_1_time.csv")

import nltk
cmu = nltk.corpus.cmudict.dict()
```

```python

def is_nan(value):
    if isinstance(value, str):
        return False
    return np.isnan(value)

def cmu_form(x, cmu):
    print(x)
    if is_nan(x):
        return None
    if x not in cmu.keys():
        return '--'
    else:
        return "-".join(cmu[x][0])

new_values = []

for index, row in into_reading.iterrows():
    
    new_value = cmu_form(row['word'], cmu)
    new_values.append(new_value)

into_reading['cmu'] = new_values

# write for inspection
into_reading.to_csv('into_reading/into_reading_sequentially_arranged_with_cmu_form.csv', index=False)
```

## Version where Into Reading is repeated 1 time

Create the datafile read in by `main()`. It needs a column called "word_raw" and a column called "count". Here for clarity, we are going to use `NA` values for `count`, which is dealt with by `main()`.

```python
data = into_reading[['word']].rename(columns = {'word': 'word_raw'})

# for this implementation, frequency is 1 for each word/row
data['count'] = 1
```

```python
data.to_csv('into_reading/into_reading_sequentially_arranged_repeated_1_time_preprocessed.csv', index=False)

main(input_file="into_reading/into_reading_sequentially_arranged_repeated_1_time_preprocessed.csv", 
     output_file="into_reading/into_reading_sequentially_arranged_repeated_1_time.pkl", 
     phonemes_path="phonreps.csv",
     cmudict_supplement="into_reading/missing_from_cmudict.json")
```
